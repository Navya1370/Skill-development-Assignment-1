```python
from google import genai
from getpass import getpass

api_key = getpass("Enter your Gemini API Key: ")
client = genai.Client(api_key=api_key)
```

```python
print("STUDENT PRODUCTIVITY ASSISTANT")
print("-" * 50)

subject = input("Enter subject: ")
deadline = input("Enter deadline: ")
topics = input("Enter topics to study: ")
hours = input("Daily study hours: ")
goal = input("Goal (Exam / Assignment / Revision): ")
```

```python
prompt = f'''
You are a smart student assistant.

Create a study plan for the student.

Subject: {subject}
Deadline: {deadline}
Topics: {topics}
Daily Study Hours: {hours}
Goal: {goal}

Instructions:
1. Make a day-wise study plan.
2. Include simple time management tips.
3. Add a short motivational message at the end.
4. Keep it clear and easy to follow.
'''
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

print("\nGENERATED STUDY PLAN")
print("-" * 50)
print(response.text)
```
